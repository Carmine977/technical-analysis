```python
from wikitables import import_tables #to download tables from the Wikipedia
import pandas as pd
```

```python
# Retrieve MIB stocks tickers from https://it.wikipedia.org/wiki/FTSE_MIB
wiki_table = import_tables('FTSE MIB', lang="it")
table = pd.read_json(wiki_table[-3].json())

MIB = pd.DataFrame(table.iloc[:,[5, 6]].values,
             index=table.iloc[:,1].values,
             columns=['Company', 'Sector'])

# save to csv file
MIB.to_csv('FMIB_tickers.csv')
MIB
```
