```python
from tradinglab.stock_manager import * 
from tradinglab.statistics import *
from tradinglab.plotting import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
T_start = "2019-01-01"
archive = "FMIB_stocks"
stocks = stock_manager(archive=archive, date_start=T_start)
data = stocks.stock_price(stock='A2A.MI')
```

```python
mav = MAV(data)
mom = momentum(data)
atr = ATR(data)
K, D = stochastic(data)
rsi = RSI(data)
```

```python
chart = trading_chart(stock_name=stocks.stock_name, figsize=(16,8))
chart.candlestick(data, width=0.6)
chart.MAV(data, window=20, alpha=0.6)
chart.hline('2019-09-16', var='Low', linestyle='--')
chart.line(date=['2019-09-17', '2019-10-29'], anchor=['Low', 'Low'], alpha=0.6)
chart.show()
```
